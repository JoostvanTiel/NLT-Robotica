# Hoofdstuk 6: Lussen
In dit hoofdstuk worden while-statements en for-statements behandeld.

## while lussen

	while conditie:
		instructies...

```python
done = False
n = 0

while not done:
	if n >= 10:
		done = True
	else: 
		print(n)
		n = n + 1
```

## for lussen

	for x in range(getal):
		instructies...

```python
for x in range(10):
	print(x)
```
